```python
# 数据处理

# 删除NaN

# 处理DataFrame
import numpy as np
import pandas as pd

NaN = np.nan
df = pd.DataFrame([[0,NaN,9,10,NaN,12],[0,2,NaN,10,NaN,12],[0,NaN,9,10,NaN,NaN]])
print(df)
print('------------')
# 删除数据,dropna删除带所有含有nan的行
print(df.dropna())
# 删除一行全为nan的
print(df.dropna(how='all'))
# 删除一列全为nan
print(df.dropna(axis=1,how='all'))

# 保留至少有thresh=n个值的行
print(df.dropna(thresh=1))
```

```python
# 数据合并,侧重于行的合并
df1 = pd.DataFrame({
    'red':[1,2,5],
    'green':[2,5,6]
},index=list('abc'))

df2 = pd.DataFrame({
    'blue':[2,3,7],
    'black':[2,3,6]
},index=list('cde'))

print(df1)
print(df2)

print('________________')
# 默认使用左连接 how = 'left'
print(df1.join(df2))
# 右合并
print(df1.join(df2,how='right'))
# 外连接
print(df1.join(df2,how='outer'))
# 内连接
print(df1.join(df2,how='inner'))
```

```python
# 时间序列
"""
D      日历日的每一天
B      工作日的每一天
H      每小时
T或min 每分钟
S      每秒
L或ms  每毫秒
U      每微秒
M      日历日的月底日期
BM     工作日的月底日期
MS     日历日的月初日期
BMS    工作日的月初日期
"""
import pandas as pd
import numpy as np

# 产生时间序列,start=时间,end=时间 默认freq='D', freq='1h20min',freq='10D'
date = pd.date_range(start='20190501',end='20190530')
print(date)
# 产生时间序列,start=时间,periods=n产生n次,freq=间隔,编号
date = pd.date_range(start='20200331',periods=1000,freq='2D')

# 时间序列获取[]
se = pd.Series(np.random.randint(0,100),index=date)
print(se)
# 通过年份获取
print(se['2021'])
# 通过月份获取
print(se['2021-4'])
# 通过日切片获取
print(se['2021-04-02':'2022-02-03'])
```

```python
# 分组聚合
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name':['BOSS','Lilei','Linei','Lilei','BOSS','Han','Zhanjj','Zhanqi'],
    'age':[2016,2016,2016,2016,2017,2017,2017,2017],
    'Salary':np.random.randn(8),
    'Bonus':[10000,20000,20000,5000,20000,30000,3000,40000]
})

# print(df)

# 根据name分组
group_by_name = df.groupby('name')
print(type(group_by_name))

# 查看分组
print(group_by_name.groups)

# 查看分组后的数量
print(group_by_name.count())

# 查看分组的情况
for name,group in group_by_name:
    print('组的名称:',name)
    print(group)
    
print('___________________')    
# 根据组的名称获取
print(group_by_name.get_group('BOSS'))

# 只获取一列
group_by_many = df['age'].groupby(df['name'])
print(group_by_many.count())

# 多重分组
group_by_name_age = df.groupby(['name','age'])
print(group_by_name_age.count())
```

```python
# 聚合函数

df = pd.DataFrame({
    'Data1':np.random.randint(0,10,5),
    'Data2':np.random.randint(10,20,5),
    'key1':list('aabba'),
    'key2':list('xyyxy')
})
print(df)
# 进行分组,只要不是数值类型的列,会自动不计入
print(df.groupby('key1').sum())
# 只需一个数据
# 先获取一列数据,再分组,计算
print(df['Data1'].groupby(df['key1']).sum())
# 先分组,取一列数据,计算
print(df.groupby('key1')['Data2'].sum())

print('____________________')
print(df.groupby('key2').mean())
```

```python
# 使用agg([函数名,函数名2,……])

df = pd.DataFrame({
    'Data1':np.random.randint(0,10,5),
    'Data2':np.random.randint(10,20,5),
    'key1':list('aabba'),
    'key2':list('xyyxy')
})

print(df.groupby('key2').agg(['sum','mean','std']))

# 可以自定义函数,传入agg(fun),改函数名称agg(('自定义函数名',原有函数名))
def peak_range(df):
    return df.max() - df.min()
print('自定义函数')
print(df.groupby('key2').agg([('范围',peak_range),'count']))
```

```python
# 扩展apply函数,apply是pandas中自由度最高的函数

df = pd.DataFrame({
    'sex':list('FFMFMMF'),
    'smoker':list('YNYNYNY'),
    'age':[21,30,17,37,30,18,26],
    'weight':[120,130,110,130,131,121,123]
})

print(df)

def bin_age(age):
    if age > 18:
        return 1
    else:
        return 0
    
#  抽烟人数大于18的
print(df['age'].apply(bin_age))

df['age'] = df['age'].apply(bin_age)

print(df)

# 取出抽烟和不抽烟体重的前二
def top(smoker,col,n=5):
    return smoker.sort_values(by=col)[-n:]

df.groupby('smoker').apply(top,col='weight',n=2)
```
