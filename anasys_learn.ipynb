```python
#绘制条形图

import matplotlib.pyplot as plt
import random

# 设置字体
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='C:/Windows/Fonts/STKAITI.TTF', size=15)


x = ['西游记','水浒传','红楼梦', '三国演义', '开国大典', '亮剑']
y = [11,13,15,10,19,17]

# 绘制x轴刻度，是一个列表，传入列表内所有参赛作为刻度; 设置刻度标签.第一个参数，确定有多少个参数以及参数的位置长度，第二个参数，具体展示
xticks_lable= ['{}'.format(i) for i in x]
plt.xticks(range(len(x)), xticks_lable, fontproperties=my_font, rotation=45)

# 绘制标题
plt.title('标题', fontproperties=my_font, color='red')

# 设置x轴标签，标签字体，倾斜45度
plt.xlabel('x轴',rotation=45,fontproperties=my_font)
# 绘制y轴
# yticks_lable = ['{}m'.format(i) for i in y1]
# plt.yticks(y, yticks_lable)

# 绘制图, label图列属性
rects = plt.bar(range(len(x)), y,width=0.5,label='电影名称')

# 添加题注
for rect in rects:
    plt.text(rect.get_x() + rect.get_width()/2, rect.get_height()+0.3,str(rect.get_height()), ha='center')

# 添加图列， 不写位置默认添加到合适位置
plt.legend(prop=my_font)

# 绘制网格
plt.grid(alpha=0.4)

# 展示
plt.show()
```

```python
# 绘制直方图
import matplotlib.pyplot as plt
import random

# 设置字体
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='C:/Windows/Fonts/STKAITI.TTF', size=15)
plt.figure(figsize=(20,10), dpi=80)

time = [random.randint(100,150) for i in range(100)]

# 绘制标题
plt.title('标题', fontproperties=my_font, color='red')

# 绘图
# 1.设置组距
distance = 2
# 2.计算组数
group_num = int((max(time)-min(time))/distance)
# 3.绘图
plt.hist(time, bins=group_num, label='电影')

# 修改刻度显示
plt.xticks(range(min(time),max(time))[::2])

# 添加图列， 不写位置默认添加到合适位置
plt.legend(prop=my_font)

# 绘制网格
plt.grid(alpha=0.4)

# 展示
plt.show()
```

```python
# 数组计算
import numpy as np
data_1 = np.arange(12).reshape(3,4)
data_2 = np.arange(0,6,0.5).reshape(3,4)
print('data1:')
print(data_1)
print('data2:')
print(data_2)

# 数组与数字计算：数组内每个数与数字进入运算（+-*/）
print('减法')
print(data_1-2)

# 数组与数组运算
# 相同形状数组运算，对应位置运算（+-*/）
sum = data_1 + data_2
print('sum:')
print(sum)
#  不同形状数组只能与相同列或行，全部运算，其余不可运算
data_plus = np.arange(4)
print('减法')
print(data_1 - data_plus)
```

```python
# 切片与索引
#  对于一维数组，连续[start：stop：span]，不连续[[数字列表]]
# 对于多维数组：两种，一直[][],另一种[一维索引,一维索引]
import numpy as np

dt = np.arange(0,24).reshape(4,6)
print('原来数组:')
print(dt)

# 获取第二行，第四行，后3列
dt_part = dt[[1,3],-3:]
print('获取第二行，第四行，后3列:')
print(dt_part)

# 修改
# 1.个体修改，根据索引修改
# 2.部分修改，可以用条件修改，[条件]   条件：<>=,  &,|,~ 与或非,数据要用（）括起来
dt[(dt>2)&~(dt>8)] = 100
print('修改数据')
print(dt)

# 三目运算np.where(condition,true,false), condition条件，返回表达式,不是修改源数组，而是返回新的数组
result = np.where(dt>99,dt+11,0)
print(result)
```

```python
# 轴axis
# 轴axis=0,1,2  二维数组只有0,1轴，方向分别沿右与下，轴，表达方向刻度，3维，0轴为横向，1,2代表二维
import numpy as np

dt = np.arange(12).reshape(3,4)
print('源数组')
print(dt)
print('0轴数组求和')
print(np.sum(dt, axis=0))
```

```python
#  添加 np.append(源数据，要添加数组,axis=)，不加轴，将源数组转化为一维数组，添加后面，加轴，直接添加后面
import numpy as np

dt = np.arange(8).reshape(2,4)
print('原数组：')
print(dt)
dt_ap = np.append(dt,[100,99])
print('添加后数组(不带轴)：')
print(dt_ap)
dt_ap_x0 = np.append(dt,[[100,99,100,99]],axis=0)
print('1轴添加数组：')
print(dt_ap_x0)
```

```python
#  数组去重
"""
numpy.unique(arr, return_index, return_inverse, return_counts)
arr:输入数组如果不是数组，则会展开；
return_index=True,返回新列表元素，在旧列表中位置，以数组保存；
retuen_inverse=True,返回旧列表中元素，在新列表中位置，以数组保存；
return_counts = True,返回去重数组中的元素数，在旧数组中出现次数,以数组保存
"""
import numpy as np
dt = np.array([1,2,3,5,6,7,1,2,3,4,6,6])
print('dt:',dt)
dt_u = np.unique(dt)
print('dt_u:',dt_u)
# 新列表元素在旧列表位置
dt_u_index, index = np.unique(dt,return_index=True)
print('dt_u_index:',dt_u_index)
print('index:',index)
# 旧列表元素在新列表位置
dt_u_inverse, inverse = np.unique(dt,return_inverse=True)
print('dt_u_inverse:',dt_u_inverse)
print('inverse:',inverse)
# 去重数组元素，在旧列表中出现次数
dt_u_counts, counts = np.unique(dt,return_counts=True)
print('dt_u_counts:',dt_u_counts)
print('counts:',counts)
```

```python
#  inf无穷大，-inf无穷小；nan缺省，与任何数值计算均为nan
# np.nan 不等，0为false， 1位true
import numpy as np

dt = np.arange(12,dtype=float).reshape(3,4)
print(dt)

dt[2,] = np.nan
print(dt)

# 统计nan数量
# np.count_nozero(),统计非零数量，0为false，本质为统计True数量
nozero_count = np.count_nonzero(dt)
print(nozero_count)
# 统计nan，np.nan != np.nan True ,即统计np.nan ！= np.nan
print(np.count_nonzero(dt != dt))

# np.isnan
print(np.isnan(dt))
#  nan替换为0
dt[np.isnan(dt)] = 0
print(dt)
```
